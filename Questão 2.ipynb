Uma companhia de gás sugeriu um modelo do tipo:

<center>$f(x) = c_1x^2 + c_2\dfrac{1}{x}$</center>

a) Aproxime pelo Método dos Mínimos Quadrados a tabela 2 pela função $f(x) = c_1x^2 + c_2\dfrac{1}{x}$. Os coeficientes devem ser exibidos com quatro casas decimais após a virgula.

b) Ache o valor para $f(x = 18)$, considere quatro cadas decimais na resposta após a virgula.

c) Plote o gráfico com a função aproximada e os pontos.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
mes = np.array([1.0000, 3.0000, 4.0000, 6.0000, 9.0000, 12.0000]) #x
consumo = np.array([10.0000, 6.5000, 5.5000, 6.0000, 10.0000, 15.0000]) #fx
```

```python
def mmq(x, y):
    g1_x = x**2
    g2_x = x**-1
    
    g1_x_g1_x = g1_x*g1_x
    g2_x_g2_x = g2_x*g2_x
    
    g1_x_g2_x = g1_x * g2_x
    
    f_g1_x = y * g1_x
    f_g2_x = y * g2_x
    
    soma_g1_x_g1_x = float(np.sum(g1_x_g1_x))
    soma_g2_x_g2_x = float(np.sum(g2_x_g2_x))
    soma_g1_x_g2_x = float(np.sum(g1_x_g2_x))
    soma_f_g1_x = float(np.sum(f_g1_x))
    soma_f_g2_x = float(np.sum(f_g2_x))
    
    Matriz_A = np.matrix([
                            [soma_g1_x_g1_x, soma_g1_x_g2_x],
                            [soma_g1_x_g2_x, soma_g2_x_g2_x]
                         ])
    
    Matriz_A = np.around(Matriz_A, 4)
    
    Vetor_B = np.array([soma_f_g1_x, soma_f_g2_x])
    
    aux_A00 = Matriz_A[0,0] / Matriz_A[0,0]
    aux_A01 = Matriz_A[0,1] / Matriz_A[0,0]
    aux_B = Vetor_B[0] / Matriz_A[0,0]
    
    aux1 = Matriz_A[1,0] - Matriz_A[1,0] * aux_A00
    aux2 = Matriz_A[1,1] - Matriz_A[1,0] * aux_A01
    aux3 = Vetor_B[1] - Matriz_A[1,0] * aux_B
    
    b = aux3 / aux2
    a = aux_B - (b * aux_A01)
    return(a, b)
```

```python
a,b = mmq(mes, consumo)
```

## Resposta da letra (a)

```python
print("O valor de c1: {:.4f}".format(a))
print("O valor de c2: {:.4f}".format(b))
```

```python
f_x = a*18**2 + b*1/18
```

```python
print("O valor para f(x = 18) é: {:.4f}".format(f_x))
```

```python
fig = plt.figure(figsize=(12,8), dpi= 100)
ax = plt.axes()
x = np.linspace(mes, 10)

g_x = a*x**2 + b/x

plt.plot(mes, consumo, 'o', color='black');
plt.plot(x, g_x, color='blue');

plt.grid()
plt.show()
```
