# Questão 5

Esse modelo é conhecido como Lotka-Volterra ou presa-predador e é dado pelas seguintes
equações:

<center>$\dfrac{dx}{dt} = x(a-by)$</center>

<center>$\dfrac{dy}{dt} = y(-c+dx)$</center>

- x: população de presas;
- y: população de predadores;
- a: taxa de crescimento efetiva da população das presas na ausência de predadores;
- b: taxa de crescimento da população de presas devido aos encontros com predadores;
- c: taxa de mortalidade da população de predadores na ausência de presas;
- d: taxa de crescimento da população de predadores devido à predação.

Simule 150 dias $(t = 150)$ deste problema. A condição inicial definida é: $x_0 = 0.5$, $y_0 = 1$, $a = 0,2$, $b = 0,7$, $c = 0,4$, $d = 0,5$, $t_0 = 0$ e $h = 0,01$. Assim, implemente a simulação utilizando com os Métodos de Euler e Runge-Kutta $4^a$ e responda quantas presas e preadores serão quando $t = 150$, exiba a resposta com dez casas decimais após a virgula. Plote um gráfico para cada Método.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x0 = 0.5
y0 = 1
a = 0.2
b = 0.7
c = 0.4
d = 0.5
t0 = 0
tf = 150
h = 0.01
```

```python
def presa_predador(x,y,a,b,c,d):
    return x * (a - b * y), y * (-c + d*x)
```

```python
#Metodo de Euler
def euler(t0, tf, x0, y0, h, a, b, c, d):
    x = [x0]
    y = [y0]
    t = [t0]

    fxy, gxy = presa_predador(x0,y0,a,b,c,d)
    x1 = x0 + h * fxy
    y1 = y0 + h * gxy
    t1 = t0 + h
    x.append(x1)
    y.append(y1)
    t.append(t1)

    n = int((tf-t0)/h)
    for i in range(1,n):
        x0 = x1
        y0 = y1
        t0 = t1
        fxy, gxy = presa_predador(x0,y0,a,b,c,d)
        x1 = x0 + h*fxy
        y1 = y0 + h*gxy
        t1 = t0 + h
        x.append(x1)
        y.append(y1)
        t.append(t1)

    return x, y, t
```

## Gráfico do Método de Euler

```python
x,y,t = euler(t0, tf, x0, y0, h, a, b, c, d)
fig = plt.figure(figsize=(12,8), dpi= 100)
plt.plot(t,x,'r-')
plt.plot(t,y,'b-')
plt.legend(["Presa","Predador"])
plt.xlabel('Tempo')
plt.ylabel('População')
plt.title('Evolução das populações no tempo')
plt.grid()
plt.show()
```

## Resultado de presas e predadores para o Método de Euler

```python
print("População de presas: {:.10f}".format(x[-1]))
print("População de predadores: {:.10f}".format(y[-1]))
```

```python
#Metodo Runge-Kutta
def rungeKutta(t0, tf, x0, y0, h, a, b, c, d):
    x = [x0]
    y = [y0]
    t = [t0]

    n = int((tf-t0)/h)
    
    for i in range(1, n + 1):
        
        fxy1, gxy1 = presa_predador(x0, y0, a, b, c, d)
        fk1 = h * fxy1
        gk1 = h * gxy1
        
        fxy2, gxy2 = presa_predador(x0 + 0.5 * fk1, y0 + 0.5 * gk1, a, b, c, d)
        fk2 = h * fxy2
        gk2 = h * gxy2
        
        fxy3, gxy3 = presa_predador(x0 + 0.5 * fk2, y0 + 0.5 * gk2, a, b, c, d)
        fk3 = h * fxy3
        gk3 = h * gxy3
        
        fxy4, gxy4 = presa_predador(x0 + fk3, y0 + gk3, a, b, c, d)
        fk4 = h * fxy4
        gk4 = h * gxy4
        
        # Atualiza o próximo valor do x e y
        x0 = x0 + (1.0 / 6.0)*(fk1 + 2.0 * fk2 + 2.0 * fk3 + fk4)
        y0 = y0 + (1.0 / 6.0)*(gk1 + 2.0 * gk2 + 2.0 * gk3 + gk4)
 
        # Atualiza o próximo valor do t
        t0 = t0 + h
        
        x.append(x0)
        y.append(y0)
        t.append(t0)
        
    return x, y, t
```

## Gráfico do Método de Runge-Kutta

```python
x,y,t = rungeKutta(t0, tf, x0, y0, h, a, b, c, d)
fig = plt.figure(figsize=(12,8), dpi= 100)
plt.plot(t,x,'r-')
plt.plot(t,y,'b-')
plt.legend(["Presa","Predador"])
plt.xlabel('Tempo')
plt.ylabel('População')
plt.title('Evolução das populações no tempo')
plt.grid()
plt.show()
```

## Resultado de presas e predadores para o Método de Runge-Kutta

```python
print("População de presas: {:.10f}".format(x[-1]))
print("População de predadores: {:.10f}".format(y[-1]))
```
